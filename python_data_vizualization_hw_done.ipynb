```python
plt.plot(x, f1, label = '$y = x^3$')
plt.plot(x, f2,  label = '$y = 2x^3$')
plt.fill_between(x, f1, f2, color='blue', alpha=0.2)
plt.legend(loc = 2)
```

На втором изображении постройте графики функции из пункта 2) и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. 

```python
x = range(-10, 11, 1)
f1 = [i**3 for i in x]
f2 = [2*i**3 for i in x]

f3 = [i**2 for i in x]
f4 = [2*i**2 for i in x]
```
